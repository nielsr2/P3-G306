```python
import cv2
import numpy as np

image = cv2.imread("mikkel.jpg")

model = cv2.dnn.readNetFromCaffe("model/deploy.prototxt", 'model/weights.caffemodel')
image = cv2.imread("mikkel.jpg")
(h, w) = image.shape[:2]
blob = cv2.dnn.blobFromImage(cv2.resize(image, (320, 320)), 1.0, (320, 320), (104.0, 177.0, 123.0))

model.setInput(blob)
detections = model.forward()

# Identify each face
for i in range(0, detections.shape[2]):
        box = detections[0, 0, i, 3:7] * np.array([w, h, w, h])
        (startX, startY, endX, endY) = box.astype("int")

        confidence = detections[0, 0, i, 2]
        print(confidence)
        # If confidence > 0.5, save it as a separate file
        if (confidence > 0.5):
        #count += 1
            frame = image[startY:endY, startX:endX]
            cv2.imshow("pikkel.jpg", frame)
#            cv2.imshow('detected circles',image)
```
