```python
import numpy as np
import cv2
from matplotlib import pyplot as plt


def findEyes(inputImg):
    # multiple cascades: https://github.com/Itseez/opencv/tree/master/data/haarcascades

    #https://github.com/Itseez/opencv/blob/master/data/haarcascades/haarcascade_frontalface_default.xml
    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    #https://github.com/Itseez/opencv/blob/master/data/haarcascades/haarcascade_eye.xml
    eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')

    inputIm = cv2.imread(inputImg)

    gray = cv2.cvtColor(inputIm, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    
    eyeImages = []
    #print('faces',type(faces))
    nd = np.ndarray(1)
    
    if (type(faces) == type(nd)):
        #print('faces_size',faces.size)
        for (x,y,w,h) in faces:
            img = cv2.rectangle(inputIm,(x,y),(x+w,y+h),0)
            roi_gray = gray[y:y+h, x:x+w]
            roi_color = inputIm[y:y+h, x:x+w]
            eyes = eye_cascade.detectMultiScale(roi_gray)
            #print("eyes",eyes,type(eyes))
            if (type(eyes) == type(nd)):
                for (ex,ey,ew,eh) in eyes:
                    cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),0)
                    crop = roi_color[ey:ey+eh, ex:ex+ew]
                    plt.imshow(crop)
                    plt.show()
                    eyeImages.append(np.asarray(crop))   
            else:
                print("No eyes found")
    else:
        print("No face found")
    return eyeImages;
```

```python
import glob
from PIL import Image
import cv2 
import numpy as np
import types
import os
from matplotlib import pyplot as plt
inputFolder = "input_images/"
outputFolder = "output_images/"

inputImages=glob.glob(inputFolder + "*.j*")

for imagePath in inputImages:
    
    # if outfolder does not exist, create it
    if not (os.path.exists(outputFolder)):
            os.mkdir(outputFolder)
            print("output folder did not exist,", outputFolder, "created.")
    # read input image
#    imgIn = cv2.imread()
    print("imagePath", imagePath)

    
    # DO STUFF TO ALL IMAGES BELOW
    # *************************

    imgOut = findEyes(imagePath)
    # *************************

    #change imagePath from input folder to output folder
    imagePath = imagePath.replace(inputFolder, "")
    inputName = imagePath.replace(".jpeg", "")
    imagePath = outputFolder + inputName

    if (type(imgOut) == type(list()) and len(imgOut) > 1):
        #print("was list")
        # if output 
        eyes = iter(imgOut)# used to iterate thru an array. the output of first next() is the first element of array, and so on
        counter = 1
        #imagePath = imagePath.replace(".", "_Left.")
        #cv2.imwrite(imagePath,next(eyes))
        #imagePath = imagePath.replace("_Left.", "_Right.")
        #cv2.imwrite(imagePath,next(eyes))
        
        for i in imgOut:
            num = str(counter)
            #imagePath = imagePath.replace(".", "_"+num+".")
            outputImg = imagePath + "_"+num+".jpeg"
            counter += 1
            cv2.imwrite(outputImg,i)
    elif (type(imgOut) == type(list()) and len(imgOut) == 1):
        print("only one eye detected")
        eyes = iter(imgOut)
        outputImg = imagePath + ".jpeg"
        cv2.imwrite(outputImg,next(eyes))
    print("done")
print("done with all")
```
