```python
# libraries
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Dataset
df=pd.DataFrame({'X': range(1,101), 'Y': np.random.randn(100)*15+range(1,101), 'Z': (np.random.randn(100)*15+range(1,101))*2 })

# plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['X'], df['Y'], df['Z'], c='skyblue', s=60)
ax.view_init(30, 185)
plt.show()
```

```python

TruePositive=np.random.uniform(90,101,[40]) 
FalseNegative=100-TruePositive
FalsePositive=np.random.uniform(0,20,[40]) 

TruePositive2=np.random.uniform(80,95,[40]) 
FalseNegative2=100-TruePositive
FalsePositive2=np.random.uniform(0,25,[40]) 

TruePositive3=np.random.uniform(60,90,[40]) 
FalseNegative3=100-TruePositive
FalsePositive3=np.random.uniform(0,30,[40]) 
dataset = TruePositive,FalseNegative,FalsePositive
print(dataset)
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.gca(projection='3d')


ax.scatter(FalseNegative, FalsePositive, TruePositive, label='Data set 1')
ax.scatter(FalseNegative2, FalsePositive2, TruePositive2, c="red", label='Data set 2')
ax.scatter(FalseNegative3, FalsePositive3, TruePositive3, c="green", label='Data set 2')

ax.legend()

ax.set_xlim(99, 1)
ax.set_ylim(1, 29)
ax.set_zlim(1, 100)
ax.set_xlabel('FalseNegative')
ax.set_ylabel('FalsePositive')
ax.set_zlabel('TruePositive')
ax.view_init(elev=20., azim=20)
#ax.xaxis.set_major_formatter(PercentFormatter(100))
#ax.yaxis.set_major_formatter(PercentFormatter(100))
#ax.zaxis.set_major_formatter(PercentFormatter(100))
plt.show()
fig.savefig("3DScatterplot.pdf", bbox_inches='tight')
```

```python
fig, ax = plt.subplots()

ax.scatter(FalsePositive, TruePositive, label='Data set 1')
ax.scatter(FalsePositive2, TruePositive2, label='Data set 2', c="red")
ax.scatter(FalsePositive3, TruePositive3, label='Data set 3', c="green")

ax.legend()
ax.set_xlim(0, 30)
ax.set_ylim(0, 100)
ax.set_xlabel('FalsePositive')
ax.set_ylabel('TruePositive')
ax.grid(True)

plt.show()
fig.savefig("2DScatterplot.pdf", bbox_inches='tight')
```
