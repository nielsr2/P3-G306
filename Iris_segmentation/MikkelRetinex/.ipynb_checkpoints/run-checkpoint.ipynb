```python
import sys
import os

import cv2
import json

import retinex

from matplotlib import pyplot as plt

data_path = 'data'
img_list = os.listdir(data_path)
if len(img_list) == 0:
    print ('Data directory is empty.')
    exit()

with open('config.json', 'r') as f:
    config = json.load(f)

for img_name in img_list:
    if img_name == '.gitkeep':
        continue
    
    img = cv2.imread(os.path.join(data_path, img_name))
    
    img_ssr = retinex.singleScaleRetinex(
    img,
    1
    )

    img_msrcr = retinex.MSRCR(
        img,
        config['sigma_list'],
        config['G'],
        config['b'],
        config['alpha'],
        config['beta'],
        config['low_clip'],
        config['high_clip']
    )
   
    img_amsrcr = retinex.automatedMSRCR(
        img,
        config['sigma_list']
    )

    img_msrcp = retinex.MSRCP(
        img,
        config['sigma_list'],
        config['low_clip'],
        config['high_clip']        
    )    

    shape = img.shape
    
    img=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
    plt.title("Original image")
    plt.imshow(img)
    plt.show()
    
    img_ssr=cv2.cvtColor(img_ssr,cv2.COLOR_RGB2BGR)
    plt.title("SSR")
    plt.imshow(img_ssr)
    plt.show()
    
    img_msrcr=cv2.cvtColor(img_msrcr,cv2.COLOR_RGB2BGR)
    plt.title("Retinex")
    plt.imshow(img_msrcr)
    plt.show()
    
    img_amsrcr=cv2.cvtColor(img_amsrcr,cv2.COLOR_RGB2BGR)
    plt.title("Automated retinex")
    plt.imshow(img_amsrcr)
    plt.show()
    
    img_msrcp=cv2.cvtColor(img_msrcp,cv2.COLOR_RGB2BGR)
    plt.title("MSRCP")
    plt.imshow(img_msrcp)
    plt.show()    
    
    #cv2.imshow('MSRCP', img_msrcp)
    #cv2.waitKey()

```
