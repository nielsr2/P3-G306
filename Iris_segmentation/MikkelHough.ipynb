```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


# Load an color image in grayscale
imgRGB = cv2.imread('eyeOpenCrop.jpg',1)

imgGrey = cv2.imread('eyeOpenCrop.jpg',0  )

#img = cv2.medianBlur(img,5)






imgRGB=cv2.cvtColor(imgRGB,cv2.COLOR_RGB2BGR)
plt.imshow(imgRGB)
plt.show()

cimg=cv2.cvtColor(imgGrey,cv2.COLOR_GRAY2BGR )
plt.imshow(cimg)
plt.show()
```

```python
cimgBlur = cv2.medianBlur(imgGrey,17)
cimg = cv2.cvtColor(cimgBlur,cv2.COLOR_GRAY2BGR)
circles = cv2.HoughCircles(cimgBlur,cv2.HOUGH_GRADIENT,1,20,param1=50,param2=30,minRadius=0,maxRadius=0)

circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    # draw the outer circle
    cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)
    # draw the center of the circle
    cv2.circle(cimg,(i[0],i[1]),2,(0,0,255),3)
    


plt.imshow(cimg)
plt.show()
```

```python
imgGreyBlur = cv2.medianBlur(imgGrey ,)
ret,th1 = cv2.threshold(imgGreyBlur,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(imgGreyBlur,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,2)
th3 = cv2.adaptiveThreshold(imgGreyBlur,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)
titles = ['Original Image', 'Global Thresholding (v = 127)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [imgGreyBlur, th1, th2, th3]
for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
for i in range(4):
    circles = cv2.HoughCircles(images[i],cv2.HOUGH_GRADIENT,1,20,param1=50,param2=30,minRadius=0,maxRadius=0)

    circles = np.uint16(np.around(circles))
    for int i in circles[0,20]:
        # draw the outer circle
        cv2.circle(images[i],(i[0],i[1]),i[2],(0,255,0),2)
        # draw the center of the circle
        cv2.circle(images[i],(i[0],i[1]),2,(0,0,255),3)


    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
    plt.show()
```
