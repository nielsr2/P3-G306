```python
import numpy as np
import cv2 
from matplotlib import pyplot as plt
import Histogram
import ThreeD_scatterplot,RGBToHSV

org_image =cv2.imread('color.jpg')

img =cv2.imread('color.jpg', cv2.IMREAD_GRAYSCALE)
#Binary image
_, mask =cv2.threshold(img,100,255,cv2.THRESH_BINARY_INV)

#Size of kernal 
kernal = np.ones((5,5),np.uint8)

#Eroision
eroison=cv2.erode(mask,kernal,iterations=2)
#Dilation
dilation = cv2.dilate(mask, kernal, iterations=2)
#Opening = eroison + dilation 
opening =cv2.morphologyEx(mask,cv2.MORPH_OPEN, kernal)
#Closing = dilation + eroison
closing =cv2.morphologyEx(mask,cv2.MORPH_CLOSE, kernal)

test = Histogram.Histogram(org_image)
test2= ThreeD_scatterplot.scatter(org_image)
test3= RGBToHSV.rgbToHSV(org_image)




titles=['image','mask','dilation','eroison','opening','closing',]
images=[img, mask, dilation,eroison,opening,closing,]

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()


```
